# Sentiment Analysis
Adapted from https://github.com/adashofdata/nlp-in-python-tutorial/blob/master/3-Sentiment-Analysis.ipynb

**Note:** If Wikipedia articles follow the NPOV guidelines, then everything should be neutral

```python
import pandas as pd

data = pd.read_pickle('corpus.pkl')
data
```

# Textblob sentimemt
The **polarity** score ranged from -1 (very negative) to 1 (very positive). This is calculated simply by averaging the positive or negative sentiment assigned to specific words.

The **subjectivity** score tells whether this is more or less opinionated. It ranges from 0 (totally objective and factual) to 1 (mostly opinion). If Wikipedia lives up to its of a Neutral Point of View (NPOV), then articles should have a score close to 0.

```python
from textblob import TextBlob

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

data['polarity'] = data['Text'].apply(pol)
data['subjectivity'] = data['Text'].apply(sub)
data
```

## Interesting results?
- Most subjectivity scores are between 0.35 and 0.4 (Is that "neutral" enough for NPOV?
- Pages for women writers have slightly higher subjectivity scores (Barbauld, Wollstonecraft, E. B. Browning, Woolf)

# View those results on a scatter plot
Note: be sure to check the axes, to see how meaningful the visualization is.

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 8]

for index, article in enumerate(data.index):
    x = data.polarity.loc[article]
    y = data.subjectivity.loc[article]
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001, data['Page_Title'][index], fontsize=10) 
    
plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)

plt.show()
```

# Sentiment analysis by parts of the article
I split each article into four parts, for convenience: ideally, we would divide each one into sections based on the section headers.

```python
import numpy as np
import math

def split_text(text, n=4):
    '''Takes in a string of text and splits into n equal parts, with a default of 4 equal parts.'''

    # Calculate length of text, the size of each chunk of text and the starting points of each chunk of text
    length = len(text)
    size = math.floor(length / n)
    start = np.arange(0, length, size)
    
    # Pull out equally sized pieces of text and put it into a list
    split_list = []
    for piece in range(n):
        split_list.append(text[start[piece]:start[piece]+size])
    return split_list
```

```python
# A list with the separate pieces of articles
list_pieces = []
for t in data.Text:
    split = split_text(t)
    list_pieces.append(split)
```

```python
#Getting the polarity for each part
article_polarity = []
for lp in list_pieces:
    polarity_piece = []
    for p in lp:
        polarity_piece.append(TextBlob(p).sentiment.polarity)
    article_polarity.append(polarity_piece)
```

# Plot the articles

To make it meaningful, would need to toggle back and forth between the graph and the article, and see which part is subjective.

```python
plt.rcParams['figure.figsize'] = [16, 12]

for index, article in enumerate(data.index):    
    plt.subplot(5, 4, index+1)
    plt.plot(article_polarity[index])
    plt.plot(np.arange(0,4), np.zeros(4))
    plt.title(data['Page_Title'][index])
    plt.ylim(ymin=-.2, ymax=.3)
    
plt.show()
```
